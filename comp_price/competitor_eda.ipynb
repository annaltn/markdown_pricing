+ which parts of our sales are more sensitive with competitor lowering price?
+ which competitor we are sensitive too?

### Helpers

```python
def add_price_diff(df):
    df['price_diff'] = df['rc_online_price'] - df['online_price']
    df['price_diff_in_percent'] = 100 * (df['rc_online_price'] - df['online_price'])/df['online_price']
    return df

def flatten(df):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
    return df

def agg_by_week(df):
    by_week = df.groupby(['sku_config_id', 'week', 'Competitor'])
    week_df = by_week.agg({'n_sold': sum, 'rc_online_price': np.median, 
                           'online_price': np.median})
    return week_df.reset_index().rename(columns={'rc_online_price': 'median_rc_online_price',
                                              'online_price': 'median_online_price'})

def agg_by_month(df):
    by_month = df.groupby(['sku_config_id', 'month', 'Competitor'])
    month_df = by_month.agg({'n_sold': sum, 'rc_price_in_sgd': np.median, 
                           'price_in_sgd': np.median})

    month_df = month_df.reset_index().rename(columns={'rc_price_in_sgd': 'median_rc_price_in_sgd',
                                              'price_in_sgd': 'median_price_in_sgd'})
    return month_df

def shorten(hier_cat, to_remove='Home > Fashion > Apparel'):
    return hier_cat.replace(to_remove, '').replace(' > ', '_')

def get_data(categories, df):
    topk_cat_df = df.loc[df.rc_categories_norm.isin(categories)]
    topk_cat_df['short_cat'] = topk_cat_df['rc_categories_norm'].apply(shorten, to_remove='Home > Fashion > Apparel > ')
    topk_cat_df = topk_cat_df.query('price_diff_in_percent < 100')
    return topk_cat_df
```

```python
def to_country(country_code):
    return {
        'id': 'Indonesia',
        'my': 'Malaysia',
        'sg': 'Singapore',
        'hk': 'HongKong',
        'tw': 'Taiwan',
        'ph': 'Philipines',
        'au': 'Australia'
    }[country_code]
```

### Load data

```python
# cols to be loaded
sku_cols = ['rc_sku_config_id', 'rc_product_name']
date_cols = ['snapshot_date']
price_cols = ['rc_online_price', 'rc_black_price', 'online_price', 'black_price', 
              'rc_currency', 'currency']
host_cols = [ 'rc_host_name', 'host_name']
cat_cols = ['rc_categories_norm', 'categories_norm']
brand_cols = ['rc_brand_norm', 'brand_norm']
usecols = sku_cols + date_cols + price_cols + host_cols + cat_cols + brand_cols

def load_competitor_data(country_code='id', start_date='2018-05-01'):
    country = to_country(country_code)
    print('Loading data of country {} starting from date {}...'.format(country, start_date))
    fname = '{}_competitors.csv'.format(country_code)
    res = pd.read_csv('../data/' + fname, usecols=usecols, parse_dates=['snapshot_date']).dropna()
    return res.loc[res.snapshot_date >= pd.to_datetime(start_date)].reset_index()
```

### Currency converter

```python
def ex_rate(currency):
    return {
        'sgd': 1,
        'hkd': 0.18,
        'myr': 0.33,
        'idr': 0.000091,
        'twd': 0.045,
        'php': 0.025,
        'aud': 0.98
    }[currency.lower()]

def to_sgd(row):
    return row['online_price'] * ex_rate(row['currency'])
```

```python
def prices_to_sgd(df):
    print('Converting currencies...')
    df['price_in_sgd'] = df[['online_price', 'currency']].apply(to_sgd, axis='columns')
    rc_currency = df.loc[0, 'rc_currency']
    df['rc_price_in_sgd'] = df['rc_online_price'] * ex_rate(rc_currency)
    return df
```

### Compute price difference

```python
def add_price_diff(df):
    print('Computing price difference...')
    df['price_diff'] = df['rc_price_in_sgd'] - df['price_in_sgd']
    df['price_diff_in_percent'] = 100 * df['price_diff']/df['price_in_sgd']
    return df
```

### Distribution of price difference by competitors

```python
def to_competitor(host_name, country_code='id'):
    #     remove country code from host name and capitalize to get competitor name
    competitor = host_name.replace('-{}'.format(country_code), '').upper()
    return str.strip(competitor)
```

```python
def dist_price_diff(df, cat_var='Competitor', country_code='id', start_date='2018-05-01'):
    
    print('Plotting price difference...')
    ax = sns.violinplot(x=cat_var, y='price_diff_in_percent', 
                        data=df.query('price_diff_in_percent <= 100') 
                       )

    country = to_country(country_code)
    title='Price difference with {} competitors (data from {})'.format(country, start_date)
    ax.set_title(title, fontsize=16)
    _ = ax.set_xlabel(cat_var, fontdict=fontdict) 
    _ = ax.set_ylabel(r'Price diff (in $\%$)', fontdict=fontdict)

    # add horizon line marking price matches
    _ = ax.axhline(y=0, linestyle='--', color='r')
    
    return ax
```

### Put all together

```python
sns.set_style('darkgrid')

def plot_price_diff_dist(competitor_df, country_code='id', start_date='2018-05-01'):
    
    competitor_df = prices_to_sgd(competitor_df)
    competitor_df = add_price_diff(competitor_df)

    # exclude outliers with too large price diff
    df = competitor_df.query('price_diff_in_percent < 1000')
    # get competitor from host name
    df['Competitor'] = df['host_name'].apply(to_competitor, country_code=country_code)

    # plot
    fig = plt.figure(figsize=(15, 5))
    ax = dist_price_diff(df=df, cat_var='Competitor', 
                         country_code=country_code,
                        start_date=start_date)

    # save plot
    fname = '{}_price_diff.pdf'.format(country_code)
    plt.savefig('plot/' + fname)
    return fig
```

```python
# finer control via JointGrid class
from scipy import stats
def plot_sale_vs_price(competitor, df):

    sub_df = df.query('Competitor == "{}"'.format(competitor.upper()))
    g = sns.JointGrid(x='median_rc_price_in_sgd', y='n_sold', 
                      data=sub_df)

    g = g.plot_joint(plt.scatter, color='b', s=40, edgecolor='white')
    g = g.plot_marginals(sns.distplot, kde=False, color='b')
    _ = g.set_axis_labels(xlabel='Median RC price in SGD', ylabel='Monthly sale', fontdict=fontdict)

    # add a statistic to summarize relationship bw 2 variables
    g = g.annotate(stats.pearsonr, fontsize=12)
    return g
```

```python
# add price diff
month_df['price_diff'] = month_df['median_rc_online_price'] - month_df['median_online_price']
month_df['price_diff_in_percent'] = 100*month_df['price_diff']/month_df['median_online_price']
month_df['price_status'] = month_df['price_diff'].apply(get_status)
```

```python
competitors = list(month_df.Competitor.unique())
half = int(len(competitors)/2)
price_status = ['higher', 'match', 'lower']

def plot_sale_vs_price_status(month_df):
    
    def plot_to(ax, competitor):
        
        qq = 'Competitor == "{}"'.format(competitor)
        sns.violinplot(x='price_status', y='n_sold', 
                        data=month_df.query(qq), 
                        order=price_status,
                       ax=ax
                       )

        # adjust aesthetics        
        _ = ax.set_title('against {}'.format(competitor), fontsize=14)
        _ = ax.set_ylabel('')
        _ = ax.set_xticklabels(labels=price_status, fontsize=12)
        return ax
    
    fig, axes = plt.subplots(2, 4, figsize=(18, 10), sharex=True)
    fig.suptitle('Sale vs. Price status', fontsize=16)
    fontdict = {'weight' : 'bold',
            'size'   : 14,
           }
    plt.xlabel('Price status', fontdict=fontdict)

    
    for i, competitor in enumerate(competitors[:half]):  
        axes[0][i] = plot_to(axes[0][i], competitor)
        axes[0][i].set_xlabel('')
    
    for i, competitor in enumerate(competitors[half: len(competitors)]):
        axes[1][i] = plot_to(axes[1][i], competitor)
        _ = axes[1][i].set_xlabel('Price status', fontdict=fontdict)
    
    axes[0][0].set_ylabel('Monthly sale', fontdict=fontdict)
    axes[1][0].set_ylabel('Monthly sale', fontdict=fontdict)
    
    return fig
```

```python
fig = plot_sale_vs_price_status(month_df)
plt.savefig('sale_vs_price.pdf')
```

__Conclusion:__ NO clear pattern. The reason is there are too many other variables affecting sales, beside price. Thus, we need to eliminate effect from other variables. In other words, when "all else are the same" then we may have a clearer picture on relationship between sale and price status.

To achieve "all else are the same", we need to limit to looking at one SKU at a time. Then the question is which SKUs to look at?

Criteria for such an SKU (ordered descendingly by importance): 
+ have on average __2 price status__ per competitor
+ have positive sale, larger should be better
+ covered by several competitors

### Bin price diff by competitor

```python
def mk_labels(quartiles):
    labels = []
    for i in range(0, len(quartiles) - 1):
        lb, ub = quartiles[i], quartiles[i+1]
        labels += ['(' + str(lb) + ', ' + str(ub) + ']']
    
    return labels
```

```python
def cut2bins(values):
    quartiles = np.round(np.percentile(values, [0, 25, 50, 75, 100]), 1)
    return pd.cut(values, bins=quartiles, include_lowest=True,
                  labels=mk_labels(quartiles))
```

```python
def bin_price_diff(competitor, df):
    print('Bin price diff of competitor {}'.format(competitor))
    sub_df = df.loc[df.host_name == competitor].copy()
    sub_df['bin_of_price_diff'] = cut2bins(sub_df['price_diff_in_percent'])
    return sub_df
```
