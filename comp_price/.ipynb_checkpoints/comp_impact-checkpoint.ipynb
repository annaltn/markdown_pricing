# Problem definition

We are working in the context of competitive pricing. Thus, for each SKU on a given date $d$, we assign a _price status_:

+ __lower__ if our price for the SKU on date $d$ is lower than all competitors. In other words, our price is smaller than the __minimum price among competitors__, denoted as $m$. 
+ __matching__ if our price equals to $m$
+ __higher__ if our price is larger than $m$

For convenient, we use the name __competitor price__ for the _minimum price_ among competitors.

__Questions:__ 
+ In general, what happen to our sales if our price is lower/matching/higher than competitor price?

For a given SKU, assume that our price is not changed.
+ If competitor price gets lower/higher than us, what impact will these have on our sales?
+ How much lower/higher will trigger the reduction/increase in sale?
+ Can we quantify the impact?

__Note on to-do:__

## Helpers

### Utilities

```python
def flatten(df):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
    return df

def shorten(hier_cat, to_remove='Home > Fashion > Apparel'):
    return hier_cat.replace(to_remove, '').replace(' > ', '_')

def get_data(categories, df):
    topk_cat_df = df.loc[df.rc_categories_norm.isin(categories)]
    topk_cat_df['short_cat'] = topk_cat_df['rc_categories_norm'].apply(shorten, to_remove='Home > Fashion > Apparel > ')
    topk_cat_df = topk_cat_df.query('price_diff_in_percent < 100')
    return topk_cat_df

def get_bounds(i, width):
    lb = width*i; ub = width*(i+1)
    return (lb, ub)

def nunique(arr):
    return len(np.unique(arr))
```

### Compute difference against competitor price

```python
def add_min_competitor_price(df):
    min_price_df = df.groupby(['sku_config_id', 'date_time']).agg({'price_in_sgd': min}).add_prefix('min_').reset_index()
    return pd.merge(df, min_price_df)

def add_price_diff(df):
    df['price_diff'] = df['rc_price_in_sgd'] - df['min_price_in_sgd']
    df['price_diff_in_percent'] = 100*df['price_diff']/df['min_price_in_sgd']
    df['price_status'] = df['price_diff'].apply(get_status)
    return df
```

```python
def dist_price_diff(df, cat_var='Competitor', title='Price difference with Indonesian competitors'):
    
    ax = sns.violinplot(x=cat_var, y='price_diff_in_percent', 
                        data=df.query('price_diff_in_percent <= 100') 
                       )

    ax.set_title(title, fontsize=16)
    _ = ax.set_xlabel(cat_var, fontdict=fontdict) 
    _ = ax.set_ylabel(r'Price diff (in $\%$)', fontdict=fontdict)

    # add horizon line marking price matches
    _ = ax.axhline(y=0, linestyle='--', color='r')
    
    return ax

def time_agg(df, time_granularity='week'):
    #     though we actually agg sales and price, we still keep their original column name 
    # s.t. we can re-use the names in computing price diff
    
    keys = ['sku_config_id'] + [time_granularity]
    cols = keys + ['n_sold', 'price_diff', 'price_diff_in_percent', 'rc_price_in_sgd', 'min_price_in_sgd']
    res = df[cols].drop_duplicates().groupby(keys).agg(
        {'n_sold': sum, 
         'price_diff': np.mean,
         'price_diff_in_percent': np.mean,
         'rc_price_in_sgd': np.mean,
         'min_price_in_sgd': np.mean
        }).reset_index()
    
    return res
```

```python
# mark when our prices are higher/lower/equal to competitors
def get_status(diff):
    if diff < 0:
        return 'lower'
    if diff == 0:
        return 'match'
    if diff > 0:
        return 'higher'
```

### Plot sale time series

```python
def sale_ts(df, time_granularity='week', factor=None):
    '''
    factor: additional factor that we want to analyze
    '''
    sns.set_style('darkgrid')
    grid = sns.relplot(x=time_granularity, y='n_sold', data=df, 
                       kind='line',
                       row=factor,
                       hue='price_status', hue_order=['lower', 'match', 'higher'], 
#                        ci=None,
                       aspect=2,
                      )

    grid = grid.set_xlabels('{}'.format(time_granularity) , fontdict=fontdict)
    grid = grid.set_ylabels('Sale', fontdict=fontdict)
    return grid
```

```python
sub_size = 14
fig_title_size = 16
```

```python
def bin_price_diff(df):
    df['bin_price_diff'] = pd.cut(df['price_diff'], 
                                    bins=10,
                                    right=False,
                                    precision=0
                                   )
    df['bin_price_diff'].cat.remove_unused_categories()
    return df

def vplot(df, ax, max_sale=20, time_granularity='weekly',
          title='RC prices are lower than competitors'):
    
    if max_sale:
        data = df.loc[df.bin_price_diff.notnull()].query('n_sold < {}'.format(max_sale)) 
    else:
        data = df.loc[df.bin_price_diff.notnull()]
    
    sns.violinplot(x='bin_price_diff', y='n_sold', data=data, ax=ax)
    # _ = ax.set_yscale('log', basey=10)
    ax.axhline(y=1, color='r', linestyle='--')
    
    _ = ax.set_title(title, fontsize=sub_size)
    _ = ax.set_ylabel('{} sale'.format(time_granularity.title()), fontdict=fontdict)
    return ax

def plot_vs_price_diff(df, time_granularity='weekly', max_sale=20, title=None):
    '''
    The DF is assumed to have only one price bin
    '''
    price_bin = list(df['price_bin'])[0]
    if not title:
        title = 'Sale vs. price difference (SKUs with price in {} SGD)'.format(price_bin) + '\n(red line represents 1 sale)'
    
    def plot_low_bin(ax, lower_df):
        ax = vplot(lower_df, 
                    ax=ax, 
                    max_sale=max_sale,
                    time_granularity=time_granularity,
                    title='RC prices are lower than competitors'
                   )
        ax.set_xlabel('')
        return ax
    
    def plot_high_bin(ax, higher_df):
        ax = vplot(higher_df, 
                    ax=ax, 
                    max_sale=max_sale,
                    time_granularity=time_granularity, 
                    title='RC prices are higher than competitors'
                   )
        _ = ax.set_xlabel('Price difference (SGD)', fontdict=fontdict)
        return ax
    
    lower_df = df.query('price_diff < 0')
    higher_df = df.query('price_diff > 0')
    
    if (lower_df.empty) and (higher_df.empty):
        print('Nothing to plot!')
        
    if (not lower_df.empty) and (not higher_df.empty):
        fig, axes = plt.subplots(2, 1, figsize=(15, 10))
        fig.suptitle(title, fontsize=fig_title_size)
        
        lower_df = bin_price_diff(lower_df)
        axes[0] = plot_low_bin(axes[0], lower_df)
        
        higher_df = bin_price_diff(higher_df)
        axes[1] = plot_high_bin(axes[1], higher_df)
        
    if (not higher_df.empty) and (lower_df.empty):
        print('We do not have prices lower than competitors')
        fig, ax = plt.subplots(1, 1, figsize=(15, 5)) 
        fig.suptitle(title, fontsize=fig_title_size)
        higher_df = bin_price_diff(higher_df)
        ax = plot_high_bin(ax, higher_df)
    if (not lower_df.empty) and (higher_df.empty):
        print('We do not have prices higher than competitors')
        fig, ax = plt.subplots(1, 1, figsize=(15, 5))
        fig.suptitle(title, fontsize=fig_title_size)
        lower_df = bin_price_diff(lower_df)
        ax = plot_low_bin(ax, lower_df)
    
    
    return fig
```

```python
fig = plot_vs_price_diff(df=weekly_df,  title='Sale vs. price difference')
```

### Micro look at SKU level

When our price is not changed, but competitor price manages to get lower than us, what will happen to our sales?

```python
sns.set_style('dark')
def plot_on_dual_axis(c1, c2, c3=None, df=None):
    fig, ax1 = plt.subplots(figsize=(10, 5))
    
    ax1.plot(df['date_time'], df[c1], color='b')
    ax1.set_xlabel('Date', fontdict=fontdict)
    ax1.set_ylabel(c1, fontdict=fontdict, color='b')
    ax1.tick_params('y', color='b')
    
    ax2 = ax1.twinx()
    ax2.plot(df['date_time'], df[c2], color='r')
    if c3:
        ax2.plot(df['date_time'], df[c3], color='r', linestyle='-.')
        ax2.set_ylabel('Price (SGD)', fontdict=fontdict, color='r')
    
#     ax2.set_yscale('symlog', basey=2, linthreshy=1)
#     ax2.set_ylabel(c2 + r'(in $\log_2$)', fontdict=fontdict, color='r')
#     ax2.axhline(0, color='k', linestyle='--')
    ax2.tick_params('y', color='r')
    plt.legend()
    return fig

def plot_sale_dynamics(sku='PU245AA0U3R2ID', data=None):
    qq = 'sku_config_id == "{}"'.format(sku)
    sub_df = data.query(qq).sort_values('date_time').reset_index()
    sub_df = sub_df.rename(columns={'rc_price_in_sgd': 'rc_price', 'min_price_in_sgd': 'min_competitor_price'})
    
    fig = plot_on_dual_axis('n_sold', 'rc_price', 'min_competitor_price', df=sub_df)
    product_name = sub_df.loc[0, 'rc_product_name']
    title = 'SKU: {} ({})\n'.format(sku, product_name)
    _ = fig.suptitle(title, fontsize=14)
    fig.subplots_adjust(top=0.85)
    return fig
```

```python
fig = plot_sale_dynamics(sku='PU245AA0U3R2ID', data=daily_df)
```

__Pattern:__
+ when our price starts matching lowest competitor, sales go up
+ when our price starts drop from matching to lower than competitor, sales spike up
+ when our price is flat again, sales drop back to 0
+ when we drop price further, up to 50%, sales increase painfully small.

The question is "Do we have a similar pattern for other SKUs?".

```python
fig = plot_sale_dynamics(sku='GR397AA0WFAZID', data=daily_df)
```

__Obs:__
+ first, our price is lower or match, our sales is high
+ when competitor price match then drop lower than us, our sale is dropped lower
