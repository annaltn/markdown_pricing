# Summary

This notebook aims at finding/confirming __relationships between features and sale/demand.__

## Inits

### Helpers

```python
# libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
```

```python
def pair_plot(data, figsize=None):
    copy = data.copy()
    copy.fillna(0, inplace=True)
    plt.figure(figsize=figsize)
    return sns.pairplot(copy)
```

```python
def to_thousand(series):
    return [int(x/(10**3)) for x in series]
```

### Load data

```python
country = 'id'
DAT_DIR = '/Users/gfg/data/markdown/clean/venture=zalora/{}/'.format(country)
```

```python
week_md_input = pd.read_csv(DAT_DIR + 'week_md_input.csv')
week_md_input.info()
```

```python
week_md_input.rename(columns={'total_sale': 'n_sold'}, inplace=True)
```

# Relationship between features and sale

## Price features vs. sale

```python
# price feats excluding totals
price_feats = [ff for ff in week_md_input.columns 
               if ('price' in ff) and ('total' not in ff) and ('ratio' not in ff)]
print('# price features: {}'.format(len(price_feats)))
```

```python
price_df = week_md_input[price_feats]
scaled = price_df.apply(to_thousand)
scaled = scaled.add_suffix('_in_thousand')
week_md_input = pd.concat([week_md_input, scaled], axis='columns')
```

```python
sample = week_md_input.sample(10**5, random_state=0)
```

```python
cols = list(scaled.columns) + ['n_sold']
pp = pair_plot(sample[cols], figsize=(10, 10))
```

## Discounts vs. sales

```python
discount_feats = [ff for ff in week_md_input.columns if ('discount' in ff) and ('total' not in ff)]
print('# discount features: {}'.format(len(discount_feats)))
```

```python
week_md_input[discount_feats].describe()
```

```python
sample[discount_feats] = np.round(sample[discount_feats], 0)
```

```python
sample[discount_feats].describe()
```

```python
cols = discount_feats + ['n_sold']
pp2 = pair_plot(sample[cols], figsize=(8,8))
```
