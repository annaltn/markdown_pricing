## Get data from EC2

```python
# get group names
fname = CLEAN_DIR + 'groups.csv'
groups = pd.read_csv(fname)
groups['gid'].head()
```

```python
def to_name(gid_str: str):
    parts = gid_str.replace('(', '').replace(')', '').split(',')
    return '_'.join([p.strip().replace("'", '') for p in parts])
```

```python
to_name(groups.gid[0])
```

```python
groups['name'] = groups['gid'].apply(to_name)
groups['name'].head()
```

```python
groups.to_csv(fname, index=False)
```
